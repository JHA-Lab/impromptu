```python
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_images_grid(imgs_, num_images=25):
  ncols = int(np.ceil(num_images**0.5))
  nrows = int(np.ceil(num_images / ncols))
  _, axes = plt.subplots(ncols, nrows, figsize=(nrows * 3, ncols * 3))
  axes = axes.flatten()

  for ax_i, ax in enumerate(axes):
    if ax_i < num_images:
      ax.imshow(imgs_[ax_i], cmap='Greys_r', interpolation='nearest')
      ax.set_xticks([])
      ax.set_yticks([])
    else:
      ax.axis('off')
```

```python
with h5.File('../../visual_analogy/datasets/clevr/clevr_train.h5', 'r') as f:
    images = f['train'][:]

num_analogies, num_examples, pair, height, width, channel = images.shape

print('Number of analogies: {}'.format(num_analogies))
print('Number of examples: {}'.format(num_examples))
print('Number of pairs: {}'.format(pair))
print('Image shape: {}x{}x{}'.format(height, width, channel))
```

```python

show_images_grid(images[0,:,0], 3)
show_images_grid(images[0,:,1], 3)
```
